 3. Check for outliers in 'price' and 'area' using the IQR method

```python
# Check for outliers in 'price' and 'area' using the IQR method
def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

print("\n--- Outliers in 'price' ---")
price_outliers, price_lower, price_upper = find_outliers_iqr(df, 'price')
if not price_outliers.empty:
    print(f"Found {len(price_outliers)} outliers in 'price'.")
    print(f"Lower bound: {price_lower:.2f}, Upper bound: {price_upper:.2f}")
    display(price_outliers[['price', 'area']].head())
else:
    print("No outliers found in 'price'.")

print("\n--- Outliers in 'area' ---")
area_outliers, area_lower, area_upper = find_outliers_iqr(df, 'area')
if not area_outliers.empty:
    print(f"Found {len(area_outliers)} outliers in 'area'.")
    print(f"Lower bound: {area_lower:.2f}, Upper bound: {area_upper:.2f}")
    display(area_outliers[['price', 'area']].head())
else:
    print("No outliers found in 'area'.")
```

### **Handling Outliers**

Although the request was not to remove outliers, here's an explanation of what could be done and why:

1.  **Investigation**: First, investigate the nature of the outliers. Are they data entry errors, or do they represent legitimate extreme values? For example, a very large `price` or `area` might belong to a luxury property that is genuinely distinct.

2.  **Transformation**: For skewed distributions (which `price` and `area` often are), transformations like logarithmic scaling (e.g., `np.log(df['price'])`) can reduce the impact of outliers and make the data more normally distributed for modeling purposes.

3.  **Capping/Winsorization**: This involves setting a cap on extreme values, replacing outliers with the nearest non-outlier value (e.g., replacing values above the upper bound with the upper bound itself, and values below the lower bound with the lower bound). This retains the data points but reduces their extreme influence.

4.  **Removal**: If outliers are clearly data errors or are few in number and highly influential, they might be removed. However, removal should be done cautiously as it can lead to loss of information and potentially bias the model if the outliers are representative of a real, albeit rare, phenomenon.

5.  **Robust Models**: Using models that are less sensitive to outliers, such as tree-based models (e.g., Random Forest, Gradient Boosting) or robust regression techniques, can be an alternative to modifying the data itself.

 4. Verify categorical columns only contain expected values

```python
# Verify categorical columns only contain expected values
categorical_cols = df.select_dtypes(include='object').columns

expected_values_map = {
    'mainroad': ['yes', 'no'],
    'guestroom': ['yes', 'no'],
    'basement': ['yes', 'no'],
    'hotwaterheating': ['yes', 'no'],
    'airconditioning': ['yes', 'no'],
    'prefarea': ['yes', 'no'],
    'furnishingstatus': ['furnished', 'semi-furnished', 'unfurnished']
}

found_unexpected = False
for col in categorical_cols:
    unique_values = df[col].unique()
    if col in expected_values_map:
        expected = set(expected_values_map[col])
        unexpected = [val for val in unique_values if val not in expected]
        if unexpected:
            print(f"Column '{col}' has unexpected values: {unexpected}")
            found_unexpected = True
        # else: # Uncomment for verbose output if no unexpected values are found for a column
            # print(f"Column '{col}' contains only expected values.")
    else:
        print(f"Warning: No expected values defined for column '{col}'. Unique values: {unique_values}")
        found_unexpected = True

if not found_unexpected:
    print("All categorical columns contain only expected values.")
```
